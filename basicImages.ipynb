```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
black_img= np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
cv2.rectangle(black_img,pt1=(34,10),pt2=(500,150),color=(0,255,0),thickness=1)
plt.imshow(black_img)
```

```python
cv2.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)
plt.imshow(black_img)
```

```python
cv2.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

```python
cv2.circle(black_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
```

```python
cv2.circle(black_img,center=(400,400),radius=50,color=(155,30,200),thickness=-1)
plt.imshow(black_img)
```

```python
cv2.line(black_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(black_img)
```

```python
font=cv2.FONT_HERSHEY_COMPLEX_SMALL
cv2.putText(black_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(black_img)
```

```python
#CUSTOM POLIGONS
white_img= np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(white_img)
```

```python
vertices= np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
cv2.polylines(white_img,[vertices],isClosed=True,color=(255,0,244),thickness=5)
plt.imshow(white_img)
```

```python
cv2.rectangle(white_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)
```

```python
#carica un immagine in cv2
img=cv2.imread('./duck.png')
img_correct=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
vertices= np.array([[100,300],[200,300],[800,300],[800,800]],dtype=np.int32)
cv2.polylines(img_correct,[vertices],isClosed=True,color=(55,0,244),thickness=10)
font=cv2.FONT_HERSHEY_COMPLEX_SMALL
cv2.putText(img_correct,text='hello baby',org=(100,500),fontFace=font,fontScale=4,color=(255,0,0),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(img_correct)
```

```python
#frey scale imag
img_gray=cv2.imread('./duck.png',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
#trasforma in bianco e nero
plt.imshow(img_gray,cmap='gray_r')
#trasforma in un numpy array
img_gray.shape
#voglio vedere il numpy array
img_gray

```

```python
#crea un quadrato  in un immagine nera con cv2
black_img= np.zeros(shape=(500,500,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
#disegna sulla black image un quadrato al centro
cv2.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(255,255,255),thickness=-1)
plt.imshow(black_img)
```

```python
from PIL import Image
import numpy as np

img = black_img
contenuto=[]
px = np.array(img)

for i in range(500):
   for j in range(500):
      if(px[i,j,0] == 255 & px[i,j,1] == 255 & px[i,j,2] == 255):
          contenuto.append({"x": i,"y": j},)
if contenuto:
   file_uno=open("esempio_uno.txt", "a")
   file_uno.write(str(contenuto))
   file_uno.close()  
 

```
